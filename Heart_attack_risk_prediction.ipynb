```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

```python
df = pd.read_csv('Heart Attack Data Set.csv')
df.head()
```

```python
fig,ax = plt.subplots(figsize=(10,6))
sns.heatmap(df.corr(),annot=True,fmt='.2f',ax=ax)
plt.show()
```

```python
sns.countplot(x = "sex",data=df, saturation=0.8)
plt.xticks(ticks=[0,1], labels = ["female", "male"])
plt.show()
```

```python
fig, ax = plt.subplots(ncols = 7, nrows = 2, figsize = (20, 10))
index = 0
ax = ax.flatten()

for cols, value in df.items():
    sns.boxplot(y=cols, data=df, ax=ax[index])
    index += 1
plt.tight_layout(pad = 0.5, w_pad=0.7, h_pad=5.0)
```

```python
fig = px.histogram (df, x = "thalach",  facet_row = "target")
fig.show()
```

```python
df.hist(figsize=(17,12),color='Pink')
plt.show()
```

```python
fig = px.scatter (df, x = "thalach", y = "chol", color = "target",trendline="lowess")
fig.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['age'], bins=20, kde=True, color='blue')
plt.title("Age Distribution")
plt.show()
```

```python
plt.figure(figsize=(7, 5))
sns.countplot(x='cp', data=df, palette='viridis')
plt.title("Chest Pain Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='trestbps', y='chol', hue='target', data=df, palette='coolwarm')
plt.title("Blood Pressure vs. Cholesterol")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='target', y='chol', data=df, palette='muted')
plt.xticks([0, 1], ['No Disease', 'Disease'])
plt.title("Cholesterol Levels by Heart Disease")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['thalach'], bins=20, kde=True, color='green')
plt.title("Maximum Heart Rate Distribution")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='cp', hue='target', data=df, palette='magma')
plt.title("Chest Pain Type and Heart Disease")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='exang', data=df, palette='husl')
plt.xticks([0, 1], ['No', 'Yes'])
plt.title("Exercise-Induced Angina")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(x='target', y='oldpeak', data=df, palette='Set2')
plt.xticks([0, 1], ['No Disease', 'Disease'])
plt.title("Oldpeak Levels by Heart Disease")
plt.show()
```

```python
sns.pairplot(df[['age', 'trestbps', 'chol', 'thalach', 'oldpeak', 'target']], hue='target', palette='husl')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x='sex', y='trestbps', data=df, palette='Set3')
plt.xticks([0, 1], ['Female', 'Male'])
plt.title("Resting Blood Pressure by Gender")
plt.show()
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(x='thal', data=df, palette='Blues')
plt.title("Thalassemia Type Count")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.kdeplot(data=df, x="age", y="thalach", hue="target", cmap="coolwarm", fill=True)
plt.title("KDE of Age vs Max Heart Rate by Heart Disease")
plt.show()
```
